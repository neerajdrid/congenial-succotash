```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x= np.linspace(0,10,50)
y= np.sin(x)
z= np.cos(x)
print(x, y, z)
```

Plotting **the** **graphs**

```python
plt.figure()
plt.plot(x,y)

#we use a different figure so that there are two different graphs otherwise we would get it in only one graph
plt.figure()
plt.plot(y,x)
plt.show()
```

```python
#parabola
x= np.linspace(-10,10,20)
y= x**2
plt.plot(x,y, 'r')
#'r' defined the color of the graph
plt.show()
```

```python
plt.figure()
plt.plot(x,y, 'ro')

plt.figure()
plt.plot(x,y, 'r.')

plt.figure()
plt.plot(x,y, 'r*')
```

```python
x= np.linspace(-5,5,50)
plt.plot(x, np.sin(x),'gx')
plt.plot(x, np.cos(x), 'rx')
plt.show()
```

Bar plot

```python
fig1= plt.figure()
ax= fig1.add_axes([0,0,1,1]) #add axes gives us the size of the graph only
languages = ["English", "German", "French " ]
people= [100, 50, 60 ]
ax.bar(languages, people)
plt.ylabel("People")
plt.xlabel("languages")
plt.show()

fig2= plt.figure()
ax2= fig2.add_axes([0.5,0.5,0.7,0.7])
languages = ["English", "German", "French " ]
people= [100, 50, 60 ]
ax2.bar(languages, people)
plt.ylabel("People")
plt.xlabel("languages")
plt.show()
```

Pie chart

```python
fig1= plt.figure()
ax= fig1.add_axes([0,0,1,1])
languages = ["English", "German", "French " ]
people= [100, 50, 60 ]
ax.pie(people, labels= languages, autopct= '%1.1f%%') # laern autopct properly
plt.show()
```

Scatterplot

```python
x= np.linspace(0,10,50)
y= np.sin(x)
z= np.cos(x)
fig2= plt.figure()
ax= fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color= "g")
ax.scatter(x, z, color= "b")
plt.show()
```

3D Scatterplot

```python
fig3= plt.figure()
ax= plt.axes(projection ="3d")
x= np.linspace(0,50,100)
y= np.sin(x)
z= np.cos(x)

ax.scatter(x,y,z, c= x, cmap= "Reds")
plt.xlabel("X")
plt.ylabel("Y")
#plt.zlabel("Z")
plt.show()
```
