```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Seaborn has some built in datasets

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_theme() #gives grids etc
sns.relplot(data= tips, x= 'total_bill', col= 'sex', y= "tip",hue = 'smoker', style= 'smoker', size='size')
#sns.scatterplot(data= tips, x= 'total_bill', y= "tip",hue = 'smoker', style= 'smoker', size='size')
```

```python
iris= sns.load_dataset("iris")
```

```python
iris.head()
```

```python
sns.scatterplot(x= "sepal_length", y= "petal_length", hue = "species", data= iris )
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.countplot(x= 'class',data= titanic )
```

```python
sns.countplot(x='survived', data= titanic)
```

```python
sns.countplot(x='who', data= titanic)
```

```python
sns.barplot(x='sex', y='survived', hue= 'embarked', data= titanic)
```
